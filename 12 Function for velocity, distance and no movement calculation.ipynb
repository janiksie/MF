### Write function

```python
def _add_velocity_and_distance_and_no_movement(df):
    
    # Create a subset of the original data frame that only contains the useful ty-values
    df1 = df[['idx','t', 'e', 'x', 'y', 'ta', 'ty', 'visitTime', 'engagementTime1', 'id', 'startTime']]
    df1 = df1[(df1.ty == 2) | (df1.ty == 3) | (df1.ty == 5) | (df1.ty == 6) | (df1.ty == 7)]
    df1 = df1.reset_index()
    
    # Define function to calculate the distance of one mouse stroke
    def _distance(row):
        if row.name == 0:
            return np.NaN
        elif (df1['t'][row.name]>=df1['t'][row.name-1]):
            dist = (((df1['x'][row.name]-df1['x'][row.name-1])**2+(df1['y'][row.name]-df1['y'][row.name-1])**2)**(0.5))
            return dist
        elif (df1['t'][row.name]<df1['t'][row.name-1]):
            return np.NaN
    
    # Apply distance function to data
    df1['distance'] = df1.apply(_distance, axis=1)
    
    # Add distances of single strokes together
    a = (df1.startTime.ne(df1.startTime.shift())).cumsum().fillna(df1['startTime'].iat[0])
    df1['dist_on_page'] = np.where(a.duplicated(keep='last'), np.nan, df1['distance'].groupby(a).transform('sum'))
            
    # Calculate the mean velocity of the strokes on one page
    def _mean_velocity(row):
        if not np.isnan(row['dist_on_page']):
            return ((row['dist_on_page']/row['visitTime']))
        else:
            return np.NaN
                                                                                                                      
    # Apply mean velocity function to data
    df1['mean_velocity'] = df1.apply(_mean_velocity, axis=1)
    
    # Define function to calculate velocity of one mouse stroke
    def _velocity(row):
        if row.name == 0:
            return np.NaN
        elif (df1['t'][row.name]>df1['t'][row.name-1]):
            v = (((df1['x'][row.name]-df1['x'][row.name-1])**2+(df1['y'][row.name]-df1['y'][row.name-1])**2)**(0.5))/((df1['t'][row.name]/1000)-(df1['t'][row.name-1]/1000))
            return v
        elif (df1['t'][row.name]==df1['t'][row.name-1]):
            return 0
        elif (df1['t'][row.name]<df1['t'][row.name-1]):
            return np.NaN
                                                                                                                      
    # Apply velocity function to data
    df1['velocity'] = df1.apply(_velocity, axis=1)
    
    # Calculate time with no movement
    max_velocity = 50

    def _no_movement(row):
        if row.name == 0:
            return 0
        elif (df1['startTime'][row.name]==df1['startTime'][row.name-1]):
            if row['velocity'] < max_velocity:
                t_no_move = df1['t'][row.name]-df1['t'][row.name-1]
                return t_no_move
    
    # Apply time with no movement function
    df1['no_movement'] = df1.apply(_no_movement, axis=1)
    
    # Add time per page together
    df1['time_no_movement'] = np.where(a.duplicated(keep='last'), np.nan, df1['no_movement'].groupby(a).transform('sum'))
    
    df1 = df1[df1['mean_velocity'].notnull()]
    
    # Merge the mean_velocity and distance_on_page columns to the original dataframe
    df = pd.merge(df, df1[['id', 'startTime', 'mean_velocity', 'dist_on_page', 'time_no_movement']],on=['id', 'startTime'], how='left')

    return df
```
