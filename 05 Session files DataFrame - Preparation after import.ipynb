### Generate column with the page category of a site and checking

```python
cat1_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/(bodenbelaege|terrassenbelag|garten|sauna|wand-und-decke|kinderwelt|kaminofen)/((?!/).)*$')
cat2_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/[\w-]+/((?!/).)*$')
cat3_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/[\w-]+/[\w-]+/((?!/).)*$')
ffadvisor_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/(parkett|laminat|gartenhaus)/((?!/).)*$')
vinylboden_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/vinylboden/((?!/).)*$')
cat2_filter_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/[\w-]+/((?!/).)*ffFilter((?!/).)*$')
product_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/.*-\w{2}\d{4}.*')
info_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/((?!/).)*$')
note_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/note((?!/).)*$')
acc_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/account.*')
checkout_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/checkout($|/.*)')
search_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/search.*')
home_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de(/$|/\?.*|((?!/).)*$)')
rechnung_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/(#rechnung|#vorkasse).*')
payment_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/payment_.*')
marken_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/[\w-]+/((?!/).)*$')
register_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/register.*')
```

```python
def _url_mapping_func(row):
    if bool(rechnung_pattern.match(row['uri'])):
        return('rechnung')
    elif bool(note_pattern.match(row['uri'])):
        return('notes')
    elif bool(acc_pattern.match(row['uri'])):
        return('account')
    elif bool(checkout_pattern.match(row['uri'])):
        return('checkout')
    elif bool(search_pattern.match(row['uri'])):
        return('search')
    elif bool(register_pattern.match(row['uri'])):
        return('register')
    elif bool(product_pattern.match(row['uri'])):
        return('product')
    elif bool(home_pattern.match(row['uri'])):
        return('home')
    elif bool(payment_pattern.match(row['uri'])):
        return('payment')
    elif bool(cat1_pattern.match(row['uri'])):
        return('cat1')
    elif bool(marken_pattern.match(row['uri'])):
        return('marken')
    elif bool(cat2_pattern.match(row['uri'])):
        return('cat2')
    elif bool(cat3_pattern.match(row['uri'])):
        return('cat3')
    elif bool(info_pattern.match(row['uri'])):
        return('info')
    else:
        return('not assigned')
```

```python
df_out['page_cat']= df_out.apply(_url_mapping_func, axis=1)
```

### Add column 'number of pages per session'

```python
num_pages_dict = dict(df_out.groupby('session_id')['idx'].count())
```

```python
df_out['num_pages'] = df_out['session_id'].apply(num_pages_dict.get)
```

### Return first five pages of a session

```python
def _first_five_pages(row):
    if row['match'] == False:
        return (','.join(list(df_out['page_cat'][row.name:row.name+5])))
```

```python
df_out['first_5_pages']= df_out.apply(_first_five_pages, axis=1)
```

### Return only pages of a session before checkout

```python
df_out['match']= df_out.session_id.eq(df_out.session_id.shift())
```

```python
df_out['before_checkout'] = True
```

```python
check = 0
for index, row in df_out.iterrows():
    if ((check == 0) and row['page_cat'] != 'checkout'):
        df_out.loc[index,'before_checkout'] = True
    else:
        df_out.loc[index,'before_checkout'] = False
        check = 1
        if row['match'] == False:
            check = 0
            df_out.loc[index,'before_checkout'] = True
```

### Sessions with no checkout in first rows

```python
df_out_head = df_out.groupby('session_id').head(1)
df_out_head_without_checkout = df_out_head[df_out_head['first_5_pages'].str.contains('checkout|payment')==False]
```

```python
list_sessions_without_checkout_infirstrows = list(df_out_head_without_checkout.groupby('session_id', as_index=False)['startTime'].count()['session_id'])
len(list_sessions_without_checkout_infirstrows)
```

```python
list_true = [True] * len(list_sessions_without_checkout_infirstrows)
no_checkout_dict =dict(zip(list_sessions_without_checkout_infirstrows, list_true))
df_out['no_checkout_infirstrows'] = df_out.set_index(['session_id']).index.map(no_checkout_dict.get)
df_out['no_checkout_infirstrows'] = df_out['no_checkout_infirstrows'].fillna(False)
```
