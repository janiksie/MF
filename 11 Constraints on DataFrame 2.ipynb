### Apply constraints 

#### The contraints applied to the session overview DataFrame are used to apply the contraints to the granular DataFrame

#### The contraints already include the restriction of 5 pages in one session

```python
constraint = df_pur['constraint']
```

```python
type(constraint)
```

```python
constraint_tuples = tuple(zip(df_pur['session_id'], df_pur['startTime']))
```

```python
len(constraint_tuples)
```

```python
purch_list_true = [True] * len(constraint_tuples)
```

```python
purch_session_constraint_dict =dict(zip(constraint_tuples, purch_list_true))
```

```python
purch_session_constraint_dict
```

```python
df_purch['match_constraint'] = df_purch.set_index(['id', 'startTime']).index.map(purch_session_constraint_dict.get)
```

```python
df_purch.groupby(['match_constraint'])['id'].nunique()
```

```python
df_purch = df_purch[df_purch['match_constraint']==True]
```

### Write file and save it to working directory
