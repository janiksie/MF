### Add column whether advisor page

```python
# Berater Seiten
advisor1_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/bodenbelaege/parkett/schiffsboden/((?!/).)*$')
advisor2_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/bodenbelaege/laminat/((?!/).)*$')
advisor3_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/terrassenbelag/terrassendielen/wpc-bpc-dielen/((?!/).)*$')
advisor4_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/bodenbelaege/parkett/((?!/).)*$')
advisor5_pattern = re.compile(r'(http://relaunch|https://www)\.holzprofi24\.de/garten/gartenhaus/((?!/).)*$')
```

```python
def _advisor_mapping_func(row):
    if bool(advisor1_pattern.match(row['url'])):
        return('Partly')
    elif bool(advisor2_pattern.match(row['url'])):
        return('Partly')
    elif bool(advisor3_pattern.match(row['url'])):
        return('Yes')
    elif bool(advisor4_pattern.match(row['url'])):
        return('Yes')
    elif bool(advisor5_pattern.match(row['url'])):
        return('Yes')
    else:
        return('No')
```

```python
df_purch['advisor']= df_purch.apply(_advisor_mapping_func, axis=1)
```
