### Apply constraint

#### Session must contain at least 5 pages

```python
df_pur_sessions = df_pur.groupby('session_id').head(1)
```

```python
pages = 5
```

```python
list_sessions_less_5_pages = list(df_pur_sessions[df_pur_sessions['num_pages']< pages]['session_id'])
```

```python
len(list_sessions_less_5_pages)
```

#### Session must contain at least one product page view

```python
df_pur.groupby(['page_cat', 'session_id'], as_index=False)['idx'].count().groupby('page_cat')['idx'].count().sort_values(ascending=False)
```

```python
df_product_sessions = df_pur.groupby(['page_cat', 'session_id'], as_index=False)['idx'].count()
```

```python
list_sessions_with_product_page = list(df_product_sessions[df_product_sessions['page_cat']=='product']['session_id'])
```

```python
list_sessions_with_product_page
```

```python
len(list_sessions_with_product_page)
```

#### Session cannot start with checkout or payment

```python
list_sessions_start_checkout = list(df_pur_sessions[df_pur_sessions['page_cat']==('checkout' or 'payment')]['session_id'])
```

```python
list_sessions_start_checkout
```

```python
len(list_sessions_start_checkout)
```

#### Use lists and apply them to df_pur

```python
df_pur['no_product_page'] = False
```

```python
for index, row in df_pur.iterrows():
    if row['session_id'] in list_sessions_with_product_page:
        df_pur.loc[index,'no_product_page']= False
    else:
        df_pur.loc[index,'no_product_page']= True
```

```python
df_pur['start_checkout'] = False
```

```python
for index, row in df_pur.iterrows():
    if row['session_id'] in list_sessions_start_checkout:
        df_pur.loc[index,'start_checkout']= True
    else:
        df_pur.loc[index,'start_checkout']= False
```

```python
df_pur['less_5_pages'] = False
```

```python
for index, row in df_pur.iterrows():
    if row['session_id'] in list_sessions_less_5_pages:
        df_pur.loc[index,'less_5_pages']= True
    else:
        df_pur.loc[index,'less_5_pages']= False
```

```python
df_pur['exclude'] = False
```

```python
for index, row in df_pur.iterrows():    
    if (row['less_5_pages'] or row['no_product_page'] or row['start_checkout']) == True:
        df_pur.loc[index,'exclude'] = True
    else:
        df_pur.loc[index,'exclude'] = False
```

#### Final purchase dataframe for further analyses

```python
df_purch = df_pur[df_pur['exclude']==False]
```

```python
del df_purch['no_product_page']
del df_purch['less_5_pages']
del df_purch['start_checkout']
del df_purch['exclude']
```

### Write file and save it to working directory

```python
start2=datetime.now()

df_purch.to_hdf('session_purchases_final_df_purch.h5', 'df_purch', mode='w')

end2 = datetime.now()
print(end2 - start2)
```
